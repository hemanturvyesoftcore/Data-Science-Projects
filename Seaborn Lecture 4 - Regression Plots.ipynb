```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv('TipsData.csv')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6,size=8)
```

```python
 iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
grid = sns.PairGrid(iris)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot)
grid.map_upper(plt.scatter)
```

```python
g = sns.FacetGrid(data=tips,row='time',col='smoker')
g.map(sns.distplot,'total_bill')
```
