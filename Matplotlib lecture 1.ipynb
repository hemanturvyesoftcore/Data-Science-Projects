```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

# Functional Method of plotting

```python
# functional method

plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('simple graph')
```

```python
# subplots
plt.subplot(1,2,1) # subplot(number or rows,number of columns,plot number)
plt.plot(x,y,'red')
plt.subplot(1,2,2)
plt.plot(y,x,'green')
```

# Object Oriented method

```python
fig = plt.figure()
axes = fig.add_axes([0,0.1,0.8,0.8]) # add_axes([left of axis, bottom of axis, widht of axis, height of axis])
axes.plot(x,y)
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_title('my graph')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X - axis')
axes.set_ylabel('Y - axis')
axes.set_title('Simple graph')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.4,0.4])
axes3 = fig.add_axes([0.3,0.7,0.2,0.2])
axes1.plot(x,y,'red')
axes2.plot(x,y,'green')
axes3.plot(x,y,'b')
```

```python
# Subplots
fig,axes = plt.subplots(nrows=1,ncols=2)
#for i in axes:
#    i.plot(x,y)

#or
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'green')
plt.tight_layout()
```

# Fig size , aspect ratio and DPI

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)

axes[1].plot(y,x,'red')
    
plt.tight_layout()
```

```python
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2,label='X Square')
axes.plot(x,x**3,label='X Cube')
axes.legend()
```

# Plot Appearence

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',marker='o',alpha=0.5,lw=3,ls=':',markersize=20,markerfacecolor='yellow',markeredgecolor='green')
```
