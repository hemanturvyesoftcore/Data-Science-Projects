```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
tips=pd.read_csv('TipsData.csv')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) 
# barplot(x = Categorical val, y = numerical val, data= dataset, estimator=Default Average)
```

```python
sns.barplot(x='total_bill',y='sex',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # std - standard deviation
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.countplot(x='sex',data=tips) #countplot = simply counts num of times the value has occures
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar',hue='sex')
```
